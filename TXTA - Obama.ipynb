## Imports

```python
import os
import slate3k as slate
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
import pandas as pd
```

## Preparing the data

```python
data_path=r"C:/LizardKing/Praxis/Obama speeches pdf/"
files=os.listdir(data_path)
```

```python
data={'topic':[],'date':[],'speech':[]}

for file in files:
    with open(data_path+file, 'rb') as f:
    
# Extracting the text    
        extracted_text = slate.PDF(f)
        extracted_text=str(extracted_text)
    
# Cleaning the text
        extracted_text=extracted_text.replace('AAmmeerr iiccaannRRhhee ttoorr iicc ..ccoomm','')
        extracted_text=extracted_text.replace('\\n','')
        extracted_text=extracted_text.replace('\\x0c','')
        extracted_text=extracted_text.replace('AmericanRhetoric.com','')
        
# Tokenizing the text
        tokenized_text=word_tokenize(extracted_text)
    
# Extracting the topic    
        topic=""
        for i in range(0,len(tokenized_text)):
            if tokenized_text[i-1]=='Obama' or tokenized_text[i-1]=='obama':
                while(tokenized_text[i]!='delivered' and tokenized_text[i]!='Delivered' and
                      tokenized_text[i]!='published' and tokenized_text[i]!='Published'):
                    topic=topic+tokenized_text[i]+" "
                    i=i+1
                break
        topic=topic.strip() 
        print(topic)
    
# Extracting the date
        date=""
        for i in range(0,len(tokenized_text)):
            if(tokenized_text[i-1]=='delivered' or tokenized_text[i-1]=='Delivered' or
               tokenized_text[i-1]=='published' or tokenized_text[i-1]=='Published'):
                    date=date+tokenized_text[i]+" "+date+tokenized_text[i+1]+" "+date+tokenized_text[i+2]
                    break
        date=date.strip() 
        print(date)
    
# Extracting the speech    
        speech=""
        for i in range(0,len(tokenized_text)):
            if tokenized_text[i-1]=='audio':
                for j in range(i,len(tokenized_text)):
                     speech=speech+tokenized_text[j]+" "
                break 
            
# Appending in the dictionary  
        if(topic!="" and date!="" and speech!=""):
            data['topic'].append(topic) 
            data['date'].append(date)
            data['speech'].append(speech)
    
    
```

- The first step is to extract the entire pdf document using Slate.
- Then some of the unwanted terms have been removed so that we are left with only the content we require for analysis.
- The third step in the above code is extracting the topic and on closely analyzing the pdf documents, I found that almost all of them have the word "Obama" just preceding the topic and is immediately followed by "delivered" or "published". The same has been utilized to extract the topics.
- The fourth step is to extract the date on which the speech was delivered and on analyzing the pdf documents, I found that almost all of them have the date immediately followed by terms "delivered" or "published" and every date object has length 3. The same has been utilized to extract the date.
- The fifth step is to extract the speech and it has been found from the documents that the speech is immediately followed by the term "audio" so the same has been utilized to extract the speech.
- The above steps have been iterated for all the documents.
- The final step is to append the topic, date and speech to a dictionary which would eventually be converted to a dataframe for easy analysis.

```python
data=pd.DataFrame(data)
```

## Some general insights from the speeches

```python
corpus=""
for i in data.index:
    corpus=corpus+data['speech'][i]+" "
```

```python
corpus=word_tokenize(corpus)
total_words = len(corpus)
print("Total number of words used in around 400 speeches are",total_words)
```

```python
unique_words = len(set(corpus))
print("Total number of unique words used in around 400 speeches are",unique_words)
```

```python
avg_words=total_words/data.shape[0]
print("Average number of words spoken per speech are",avg_words)
```

```python
# pretty simple function which is essentially unique_words / total_words * 100
def lexical_diversity(text_in):
    total_words = len(text_in)
    total_unique = len(set(text_in))
    lex_div = (total_unique / total_words) * 100
    return lex_div
```

```python
lex_d = lexical_diversity(corpus)
print("{:.2f}".format(lex_d),'% diversity')
```

- We see that there is 1.77 percent lexical diversity in the speeches delivered by Obama.

## Finding the most frequent nouns Obama has talked about using POS tagging

```python
text=nltk.Text(corpus)
```

```python
nltk.download('averaged_perceptron_tagger')
tags = nltk.pos_tag(text)
```

```python
tags
```

```python
counts={}
for word, tag in tags:
    if tag=='NN' or tag=='NNS' or tag=='NNP':
        if word not in counts.keys():
            counts[word]=1
        else:
            counts[word]+=1
```

- The above code prepares a dictionary with nouns as keys and their count as the values.

```python
sorted_counts = sorted(counts.items(), key=lambda x: x[1],reverse=True)
```

```python
sorted_counts
```

- The above code finds the most talked about classes of people, places, or things.
- We can see that the nouns he has talked about most are people, president, America, world, time, country, years, way, today, government, Obama, work, security, health, countries, economy and nation among many others.
